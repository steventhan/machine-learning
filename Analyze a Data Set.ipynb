## Visualize: The relationship between the proportion of black citizens and the distance to employment centers in Boston.

```python
black_proportion = (np.sqrt(a[:, 11]) / 1000) + 0.63
```

```python
distance_to_emp_centers = a[:, 7]
```

```python
plt.scatter(black_proportion, distance_to_emp_centers)
```

```python
plt.plot(np.unique(black_proportion), np.poly1d(np.polyfit(black_proportion, distance_to_emp_centers, 1))(np.unique(black_proportion)))
```

```python
plt.xlabel("Black proportion")
```

```python
plt.ylabel("Distance to employment centers")
```

```python
plt.title('Relationship between the proportion of black citizens and the distance to employment centers in Boston')
```

```python
plt.show()
```

With only the scatter plot, it wasn't very conclusive, so I tried adding a regression line, it seems there's a trend in which the larger black community located further away from employment centers.
